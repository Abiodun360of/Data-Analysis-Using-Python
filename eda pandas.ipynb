```python
# Group by 'Continent' and calculate the mean for the specified columns
group_year_bycontinent = (
    df.groupby('Continent')[['2022 Population', '2020 Population', '2015 Population', 
                             '2010 Population', '2000 Population', '1990 Population', 
                             '1980 Population', '1970 Population']]
    .mean()
    .sort_values(by='2022 Population', ascending=False)
)

# Display the result
print(group_year_bycontinent)
```

```python
group_year_bycontinent.
```

```python

group_year_bycontinent[::-1].transpose().plot(figsize=(12, 10))

# Add title and labels
plt.title('Yearly Population by Continent', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)


```

```python
import seaborn as sns
sns.scatterplot(data=group_year_bycontinent, x='Continent', y='2010 Population')

```
